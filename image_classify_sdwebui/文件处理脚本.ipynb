```python
# 重命名文件
# 指定包含要重命名文件的文件夹路径
folder_root = 'C:\\Users\\yhk\\Desktop\\Workspace\\classification-pytorch\\datasets'

for dir_name in ["训练集_train", "测试集_val"]:
    for tag_name in ["不要的", "普通", "优质", "瑕疵"]:
        folder_path = f"{folder_root}\\{dir_name}\\{tag_name}"
        # 列出文件夹中的所有文件
        files = os.listdir(folder_path)
        # 过滤出特定类型的文件，例如 .jpg 文件
        jpg_files = [f for f in files if f.endswith('.png')]
        # 遍历文件并重命名
        for i, file_name in enumerate(jpg_files, start=1):
            # 构建新文件名
            new_name = f"{dir_name}_{tag_name}_{i}.jpg"
            # 获取旧文件名和新文件名的完整路径
            old_path = os.path.join(folder_path, file_name)
            new_path = os.path.join(folder_path, new_name)
            # 重命名文件
            os.rename(old_path, new_path)

        print("Files renamed successfully.")
```

```python
# 计算图片文件md5
from collections import defaultdict
import hashlib
import os


def get_md5_from_path(file_path):
    md5 = None
    if os.path.exists(file_path):
        with open(file_path, 'rb') as f:
            md5_obj = hashlib.md5()
            md5_obj.update(f.read())
            hash_code = md5_obj.hexdigest()
            md5 = str(hash_code).lower()
    return md5


# 检查同级目录下 MD5一样的文件有那些

for root, _, files in os.walk(r'D:\my_deeplearn\image_classify_sdwebui\data\test'):
    file_md5_dict = defaultdict(list)
    for file in files:
        file_path = os.path.join(root, file)
        file_md5 = get_md5_from_path(file_path)
        if file_md5 is None:
            continue
        file_md5 = file_md5.lower()
        file_md5_dict[file_md5].append(file_path)

    # file_md5_dict 长度超过1的有哪些
    for k, v in file_md5_dict.items():
        if len(v) > 1:
            print(k, v)
            for v_item in v[1:]:
                os.remove(v_item)
```
